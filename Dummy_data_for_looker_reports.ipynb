```python
import random
import csv

# Extended list of first and last names
first_names_with_gender = {
    "Michael": "M", "Emily": "F", "Alexander": "M", "Amanda": "F", "Sophia": "F", "David": "M", "Isabella": "F",
    "Mason": "M", "Jacob": "M", "Olivia": "F", "Emma": "F", "Liam": "M", "Ava": "F", "James": "M", "William": "M",
    "Mia": "F", "Aiden": "M", "John": "M", "Grace": "F", "Elizabeth": "F", "Noah": "M", "Logan": "M", "Lucas": "M",
    "Ethan": "M", "Daniel": "M", "Matthew": "M", "Elijah": "M", "Ella": "F", "Abigail": "F", "Chloe": "F",
    "Zoey": "F", "Natalie": "F", "Hannah": "F", "Lily": "F", "Avery": "F", "Samantha": "F", "Addison": "F",
    "Victoria": "F", "Jackson": "M", "Sebastian": "M", "Jack": "M", "Gabriel": "M", "Carter": "M", "Henry": "M",
    "Julian": "M", "Levi": "M", "Wyatt": "M", "Isaac": "M", "Ryan": "M", "Landon": "M", "Caleb": "M", "Nathan": "M",
    "Luke": "M", "Christian": "M", "Hunter": "M", "Owen": "M", "Jordan": "M", "Cameron": "M", "Connor": "M",
    "Charles": "M", "Aaron": "M", "Adrian": "M", "Eli": "M", "Jeremiah": "M", "Thomas": "M", "Evan": "M",
    "Angel": "M", "Robert": "M", "Gavin": "M", "Nicholas": "M", "Dominic": "M", "Jaxon": "M", "Isaiah": "M",
    "Mila": "F", "Eleanor": "F", "Penelope": "F", "Aria": "F", "Riley": "F", "Scarlett": "F", "Layla": "F",
    "Lillian": "F", "Nora": "F", "Aurora": "F", "Savannah": "F", "Brooklyn": "F", "Bella": "F", "Claire": "F",
    "Skylar": "F", "Lucy": "F", "Paisley": "F", "Everly": "F", "Anna": "F", "Caroline": "F", "Nova": "F",
    "Genesis": "F", "Emilia": "F", "Kennedy": "F", "Madelyn": "F", "Piper": "F", "Ruby": "F", "Serenity": "F",
    "Mackenzie": "F", "Hadley": "F", "Autumn": "F", "Ariana": "F", "Sadie": "F", "Delilah": "F", "Josephine": "F",
    "Nevaeh": "F", "Quinn": "F", "Ivy": "F", "Maya": "F", "Sophie": "F", "Alice": "F", "Isabelle": "F", "Vivian": "F",
    "Gianna": "F", "Rylee": "F", "Peyton": "F", "Evelyn": "F", "Madeline": "F", "Sloane": "F", "Jade": "F",
    "Ada": "F", "Eliza": "F", "Cecilia": "F", "Mary": "F", "Margaret": "F", "Harmony": "F", "Annabelle": "F",
    "Rose": "F", "Alexandra": "F", "Jocelyn": "F", "Reagan": "F", "Valentina": "F", "Aubree": "F", "Hazel": "F",
    "Brooke": "F", "London": "F", "Genevieve": "F", "Daisy": "F", "Isla": "F", "Liliana": "F", "Eliana": "F",
    "Jordyn": "F", "Morgan": "F", "Finley": "F", "Zoe": "F", "Kaylee": "F", "Alyssa": "F", "Paislee": "F",
    "Sydney": "F", "Ryleigh": "F", "Cali": "F", "Rosalie": "F", "Emerson": "F", "Aliyah": "F", "Lydia": "F",
    "Adeline": "F", "Ximena": "F", "Isabel": "F", "Ashley": "F", "Brianna": "F", "Andrea": "F", "Bailey": "F",
    "Maria": "F", "Melanie": "F", "Iris": "F", "Rebecca": "F", "Arabella": "F", "Aubree": "F", "Giselle": "F",
    "Brielle": "F", "Valerie": "F", "Autumn": "F", "Juliana": "F", "Jayla": "F", "Sienna": "F", "Laila": "F",
    "Marley": "F"
}

last_names = [
    "Johnson", "Smith", "Jones", "Lee", "Martinez", "Rodriguez", "Miller", "Davis", "Garcia",
    "Martinez", "Anderson", "Thomas", "White", "Harris", "Lopez", "Robinson", "Clark",
    "Young", "King", "Brown", "Wilson", "Taylor", "Moore", "Jackson", "Martin", "Lee",
    "Perez", "Thompson", "Hernandez", "Walker", "Allen", "Young", "Hall", "Wright", "King",
    "Green", "Baker", "Adams", "Nelson", "Hill", "Ramirez", "Campbell", "Mitchell", "Roberts",
    "Carter", "Phillips", "Evans", "Turner", "Torres", "Parker", "Collins", "Edwards", "Stewart",
    "Flores", "Morris", "Nguyen", "Murphy", "Rivera", "Cook", "Rogers", "Morgan", "Peterson",
    "Cooper", "Reed", "Bailey", "Bell", "Gomez", "Kelly", "Howard", "Ward", "Cox", "Diaz",
    "Richardson", "Wood", "Watson", "Brooks", "Bennett", "Gray", "James", "Reyes", "Cruz",
    "Hughes", "Price", "Myers", "Long", "Foster", "Sanders", "Ross", "Morales", "Powell",
    "Sullivan", "Russell", "Ortiz", "Jenkins", "Gutierrez", "Perry", "Butler", "Barnes",
    "Fisher"
]

# Schools and counselors setup
schools = {
    31: "Chester ES",
    32: "Greenville ES",
    34: "Lebanon ES",
    36: "Riverside ES",
    40: "Salem ES",
    42: "Charlotte ES",
    51: "Oakland ES",
    52: "Springfield ES",
    70: "Bristol Cyber School",
    81: "Walker HS"
}

# Counselor assignments
counselors = {
    31: "Anderson, Chris",
    32: "Anderson, Chris",
    34: "Anderson, Chris",
    36: "Brown, Pat",
    40: "Brown, Pat",
    42: "Brown, Pat",
    51: "Davis, Jordan",
    52: "Davis, Jordan",
    70: "Miller, Alex",
    81: {
        "9": "Taylor, Morgan",
        "10_A_H": "Johnson, Casey",
        "10_I_R": "Smith, Jamie",
        "10_S_Z": "Williams, Taylor"
    }
}

# List of PK-8 school IDs
pk8_school_ids = [31, 32, 34, 36, 40, 42, 51, 52]

ethnicities = [
    "American Indian or Alaska Native", "Asian", "Black or African American",
    "Hispanic or Latino", "Native Hawaiian or Other Pacific Islander", "White",
    "Two or More Races"
]

def generate_name():
    first_name = random.choice(list(first_names_with_gender.keys()))
    gender = first_names_with_gender[first_name]

    # Select a second (middle) name with the same gender
    possible_middle_names = [name for name, g in first_names_with_gender.items() if g == gender]
    middle_name = random.choice(possible_middle_names)

    last_name = random.choice(last_names)

    return last_name, first_name, middle_name, gender

def generate_homeroom(school_id, grade_level):
    if grade_level == -1:
        # Pre-Kindergarten homerooms should be within the previously defined parameters
        school_abbr = random.choice(list(schools.keys()))
        room_number = f"{random.randint(1, 399):02d}"
    else:
        # For other grades, generate homerooms as before
        school_abbr = schools[school_id].split()[0][:2].upper()
        room_number = f"{random.randint(1, 399):03d}"
    return f"{school_abbr}{room_number}"

def generate_state_student_id():
    return str(random.randint(1000000000, 9999999999))

def generate_disability():
    # National averages: About 14% of students have disabilities, 504 plans ~2.3%
    special_education = random.random() < 0.14
    has_504 = False
    primary_disability = None
    if special_education:
        has_504 = False
        primary_disability = random.randint(1000, 9999)
    else:
        has_504 = random.random() < 0.023
    return primary_disability, special_education, has_504

def generate_random_grade_distribution(total_students, grades):
    distribution = {}
    remaining_students = total_students
    last_grade_count = random.randint(15, 70)
    for grade in grades[:-1]:
        grade_count = random.randint(max(15, last_grade_count - 70), min(70, last_grade_count + 70))
        grade_count = min(grade_count, remaining_students)
        distribution[grade] = grade_count
        remaining_students -= grade_count
        last_grade_count = grade_count
    distribution[grades[-1]] = remaining_students
    return distribution

# National average occurrence rate for each ethnicity
national_ethnicity_occurrence_rate = {
    "White": 0.60,
    "Black or African American": 0.13,
    "Hispanic or Latino": 0.18,
    "Asian": 0.06,
    "American Indian or Alaska Native": 0.01,
    "Native Hawaiian or Other Pacific Islander": 0.002,
    "Two or More Races": 0.04
}

# Initialize student list
students = []

# Distribute students
high_school_students = random.randint(1200, 1400)
cyber_school_students = random.randint(475, 550)
prek_students = random.randint(280, 300)
remaining_students = 6000 - (high_school_students + cyber_school_students + prek_students)


# Generate data for PreK students
for _ in range(prek_students):
    primary_disability, special_education, has_504 = generate_disability()
    school = random.choice(pk8_school_ids)  # Randomly assign a PK-8 school
    last_name,first_name,middle_name,gender = generate_name()
    students.append([
        school,  # Randomly assign a PK-8 school
        schools[school],  # School title
        counselors[school],  # Assign counselor based on school
        generate_homeroom(school, -1),  # Generate homeroom for PreK
        random.randint(100000, 999999),  # Student ID
        last_name,  # Last name
        first_name,  # First name
        middle_name,  # Middle name
        -1,  # Grade level for PreK
        random.choices(list(national_ethnicity_occurrence_rate.keys()),
            weights=list(national_ethnicity_occurrence_rate.values()))[0],  # Ethnicity
        has_504,  # 504 status
        primary_disability,  # Primary disability
        special_education,  # Special education status
        random.random() < 0.1,  # LEP_FLG
        random.random() < 0.03,  # HOMELESS
        None,  # LEP
        True,  # CURRENT_ENROLLMENT
        generate_state_student_id(),  # State student ID
        gender
    ])



# Randomly distribute the total number of students across grade levels while ensuring the total remains within the specified range
hs_grades_distribution = {}
hs_remaining_students = high_school_students
for grade in range(12, 8,-1):
    if grade == 12:
        # Randomly select between 30% and 36% of the remaining students for grades 10 & 11
        min_percentage = 0.20
        max_percentage = 0.25
        selected_percentage = random.uniform(min_percentage, max_percentage)
        hs_grades_distribution[grade] = int(selected_percentage * hs_remaining_students)
        hs_remaining_students -= hs_grades_distribution[grade]
    elif grade == 11:
        min_percentage = 0.30
        max_percentage = 0.40
        selected_percentage = random.uniform(min_percentage, max_percentage)
        hs_grades_distribution[grade] = int(selected_percentage * hs_remaining_students)
        hs_remaining_students -= hs_grades_distribution[grade]
    elif grade == 10:
        min_percentage = 0.45
        max_percentage = 0.50
        selected_percentage = random.uniform(min_percentage, max_percentage)
        hs_grades_distribution[grade] = int(selected_percentage * hs_remaining_students)
        hs_remaining_students -= hs_grades_distribution[grade]
    else:
        # Assign the remaining students equally among grades 10-12
        hs_grades_distribution[grade] = hs_remaining_students


hs_homerooms = []
for grade, count in hs_grades_distribution.items():
    for _ in range(count):
        homeroom = generate_homeroom(81, grade)
        hs_homerooms.append(homeroom)
        primary_disability, special_education, has_504 = generate_disability()
        counselor = counselors[81]["9"] if grade == 9 else \
            counselors[81]["10_A_H"] if "A" <= last_name <= "H" else \
            counselors[81]["10_I_R"] if "I" <= last_name <= "R" else \
            counselors[81]["10_S_Z"]
        last_name,first_name,middle_name,gender = generate_name()
        students.append([
            81, schools[81], counselor, homeroom,
            random.randint(100000, 999999), last_name,
            first_name, middle_name,
            grade, random.choices(list(national_ethnicity_occurrence_rate.keys()),
               weights=list(national_ethnicity_occurrence_rate.values()))[0], has_504, primary_disability,
            special_education, random.random() < 0.1, random.random() < 0.03,
            None, True, generate_state_student_id(),gender
        ])

# Randomly distribute the total number of students across grade levels while ensuring the total remains within the specified range
cyber_grade_distribution = {}
cyber_remaining_students = cyber_school_students
for grade in range(12, -1,-1):
    if grade in [9,10,11,12]:
        # Randomly select between 30% and 36% of the remaining students for grades 10 & 11
        min_percentage = 0.12
        max_percentage = 0.18
        selected_percentage = random.uniform(min_percentage, max_percentage)
        cyber_grade_distribution[grade] = int(selected_percentage * cyber_school_students)
        cyber_remaining_students -= cyber_grade_distribution[grade]
    elif grade in [5,6,7,8]:
        min_percentage = 0.02
        max_percentage = 0.03
        selected_percentage = random.uniform(min_percentage, max_percentage)
        cyber_grade_distribution[grade] = int(selected_percentage * cyber_school_students)
        cyber_remaining_students -= cyber_grade_distribution[grade]
    else:
        selected_percentage = .3
        cyber_grade_distribution[grade] = int(selected_percentage * cyber_remaining_students)
        cyber_remaining_students -= cyber_grade_distribution[grade]
        selected_percentage += .3

for grade, count in cyber_grade_distribution.items():
    for _ in range(count):
        primary_disability, special_education, has_504 = generate_disability()
        last_name,first_name,middle_name,gender = generate_name()
        students.append([
            70, schools[70], counselors[70], generate_homeroom(70, grade),
            random.randint(100000, 999999), last_name, first_name, middle_name,
            grade, random.choices(list(national_ethnicity_occurrence_rate.keys()),
               weights=list(national_ethnicity_occurrence_rate.values()))[0], has_504, primary_disability,
            special_education, random.random() < 0.1, random.random() < 0.03,
            None, True, generate_state_student_id(),gender
        ])

# Distribute students across K-8 schools
k8_students_distribution = remaining_students // 8
k8_school_distribution = {school: k8_students_distribution + random.randint(-35, 35) for school in list(schools.keys())[:8]}

# Generate data for K-8 students
for school, total_students in k8_school_distribution.items():
    # Calculate the number of students for each grade level
    grades = list(range(0, 9))
    grade_distribution = {grade: total_students // len(grades) for grade in grades}

    # Distribute the remaining students randomly across grades
    remaining_students = total_students - sum(grade_distribution.values())
    while remaining_students > 0:
        grade = random.choice(grades)
        grade_distribution[grade] += 1
        remaining_students -= 1

    # Generate student data for K-8 students in each school
    for grade, count in grade_distribution.items():
        for _ in range(count):
            primary_disability, special_education, has_504 = generate_disability()
            last_name,first_name,middle_name,gender = generate_name()
            students.append([
                school, schools[school], counselors[school],
                generate_homeroom(school, grade), random.randint(100000, 999999),
                last_name,first_name,middle_name, grade, random.choices(list(national_ethnicity_occurrence_rate.keys()),
                               weights=list(national_ethnicity_occurrence_rate.values()))[0],
                has_504, primary_disability, special_education,
                random.random() < 0.1, random.random() < 0.03,
                None, True, generate_state_student_id(),gender
            ])



# Write to CSV
with open('students.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow([
        "SCHOOL_ID", "SCHOOL_TITLE", "COUNSELOR", "HOME_ROOM", "STUDENT_ID",
        "LAST_NAME", "FIRST_NAME", "MIDDLE_NAME", "GRADE_LEVEL", "ETHNICITY_DESC",
        "504", "PRIMARY_DISABILITY", "SPECIAL_EDUCATION", "LEP_FLG", "HOMELESS",
        "LEP", "CURRENT_ENROLLMENT", "STATE_STUDENT_ID"
    ])
    writer.writerows(students)

# Calculate and print enrollment counts for each school
school_enrollment = {school: sum(1 for student in students if student[0] == school) for school in schools}
print("School Enrollment:")
for school, enrollment in school_enrollment.items():
    print(f"School ID {school}: {enrollment} students")

# Calculate and print enrollment counts for each grade
grade_enrollment = {grade: sum(1 for student in students if student[8] == grade) for grade in range(-1, 13)}
print("\nGrade Enrollment:")
for grade, enrollment in grade_enrollment.items():
    print(f"Grade {grade}: {enrollment} students")
```
